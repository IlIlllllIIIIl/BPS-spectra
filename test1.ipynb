```python
from star import StarSolver, save
import matplotlib.pyplot as plt
import pandas as pd
import sqlite3
import numpy as np
import seaborn as sns

r = [-30] * 4
solver = StarSolver(r)
solver.find_all()

solution = solver.get_solutions()
flatsol = []
for key, matrix in solution.items():
    for row, mat in enumerate(matrix):
        flatsol.append({"det": key, "row": row, **{f"col{j}": val for j, val in enumerate(mat)}})

conn = sqlite3.connect("star_10.db")
df = pd.DataFrame(flatsol)
df.to_sql("star_10", conn, if_exists="replace", index=False)

conn.close()

```
